## Part 2 — Static Method Utility Class (~10 minutes)

Create a class named `ArenaTools`.

This class should contain two static methods.

### `clean_action(action: str) -> str`

Returns the action:
- stripped of whitespace
- converted to lowercase

Example:

```python
ArenaTools.clean_action("  FIREBALL ")
```

should return:

```python
"fireball"
```

### `is_valid_action(action: str) -> bool`

Returns `True` if the action:
- contains only alphabetic characters
- has at least 3 characters

Returns `False` otherwise.

Examples:

```python
ArenaTools.is_valid_action("run")       # True
ArenaTools.is_valid_action("fireball")  # True
ArenaTools.is_valid_action("hi")        # False
ArenaTools.is_valid_action("fire2")     # False
```


````python
# Part 2 code here
class ArenaTools:
  @staticmethod
  def clean_action(action: str) -> str:
    return action.strip().lower()

  @staticmethod
  def is_valid_action(action: str) -> bool:
    if action.isalpha() and len(action) >= 3:
      return True
    else:
      return False
````

### Part 2 Test Cell

Run this cell after creating `ArenaTools`.


````python
print(ArenaTools.clean_action("  FIREBALL "), "expected: fireball")
print(ArenaTools.is_valid_action("run"), "expected: True")
print(ArenaTools.is_valid_action("fireball"), "expected: True")
print(ArenaTools.is_valid_action("hi"), "expected: False")
print(ArenaTools.is_valid_action("fire2"), "expected: False")
````

## Part 3 — Abstract Parent Class (~10 minutes)

Create an abstract parent class named `ArenaCharacter`.

This class represents the contract for all arena characters.

Use:

```python
from abc import ABC, abstractmethod
```

### Required Constructor

Each character should have:

- `name`
- `health`

### Required Abstract Methods

Every child class must implement:

```python
perform_action(action: str)
show_status()
```

### Class Attribute and Class Method

Add a class attribute:

```python
character_count = 0
```

Every time an `ArenaCharacter` is created, increase the count.

Add a class method:

```python
get_character_count()
```

that returns the number of characters created.


````python
# Part 3 code here
from abc import ABC, abstractmethod

class ArenaCharacter(ABC):
  character_count = 0

  def __init__(self, name, health):
    self.name = name
    self.health = health
    ArenaCharacter.character_count += 1

  @abstractmethod
  def perform_action(self, action: str):
    pass

  @abstractmethod
  def show_status(self):
    pass

  @classmethod
  def get_character_count(cls):
    return cls.character_count

````

## Part 4 — Concrete Child Class: `Wizard` (~15 minutes)

Create a child class named `Wizard` that inherits from `ArenaCharacter`.

### Required Additional Attribute

Each `Wizard` should also have:

- `mana`

### Required Methods

#### `perform_action(action: str)`

This method should:

- clean the action using `ArenaTools.clean_action()`
- validate the action using `ArenaTools.is_valid_action()`
- if the action is invalid, print an error message and return `False`
- if the action is valid and the wizard has at least 10 mana:
  - subtract 10 mana
  - print a message such as `"Merlin casts fireball!"`
  - return `True`
- if the wizard does not have enough mana:
  - print a message saying there is not enough mana
  - return `False`

#### `show_status()`

Prints the wizard's name, health, and mana.


````python
# Part 4 code here
class Wizard(ArenaCharacter):
  def __init__(self, name, health, mana):
    self.mana = mana
    super().__init__(name, health)

  def perform_action(self, action: str):
    action = ArenaTools.clean_action(action)

    if not ArenaTools.is_valid_action(action):
      print("This is not a valid action")
      return False

    elif self.mana >= 10:
      self.mana -= 10
      print(f"{self.name} did {action}")
      return True

    else:
      print("There is not enough mana")
      return False

  def show_status(self):
    print(f"name: {self.name}, health: {self.health}, mana: {self.mana}")

````

## Part 5 — Concrete Child Class: `Warrior` (~15 minutes)

Create another child class named `Warrior` that inherits from `ArenaCharacter`.

### Required Additional Attribute

Each `Warrior` should also have:

- `stamina`

### Required Methods

#### `perform_action(action: str)`

This method should:

- clean the action using `ArenaTools.clean_action()`
- validate the action using `ArenaTools.is_valid_action()`
- if the action is invalid, print an error message and return `False`
- if the action is valid and the warrior has at least 5 stamina:
  - subtract 5 stamina
  - print a message such as `"Athena performs slash!"`
  - return `True`
- if the warrior does not have enough stamina:
  - print a message saying there is not enough stamina
  - return `False`

#### `show_status()`

Prints the warrior's name, health, and stamina.


````python
# Part 5 code here
class Warrior(ArenaCharacter):
  def __init__(self, name, health, stamina):
    super().__init__(name, health)
    self.stamina = stamina

  def perform_action(self, action: str):
    action = ArenaTools.clean_action(action)
    valid = ArenaTools.is_valid_action(action)
    if not valid:
      print("Action is not valid")
      return False
    elif self.stamina >= 5:
      self.stamina -= 5
      print(f"{self.name} performs {action}")
      return True
    else:
      print("There is not enough stamina")
      return False

  def show_status(self):
    print(f"name: {self.name}, health: {self.health}, stamina: {self.stamina}")
````

In Python, a class can inherit from more than one parent class. This is called **multiple inheritance**.

One common use of multiple inheritance is a **mixin**.

Answers:
1) This class inherits methods and attributes from two parent classes: Wizard and FlyinMixin
2) The perform_action and show_status behaviors came from Wizard, as well as the constructor that initialized name, health, and mana
3) The take_off, land, and show_flight_status methods came form FlyingMixin, as well as the is_flying attribute.
4) Mixins can be useful when you need to add a specific behavior or set of behaviors to an existing child class.

````python
class FlyingMixin:
  def __init__(self):
    self.is_flying = False

  def take_off(self):
    self.is_flying = True
    print("Taking off!")

  def land(self):
    self.is_flying = False
    print("Landing!")

  def show_flight_status(self):
    if self.is_flying:
      print("Flying")
    else:
      print("Not Flying")
````

````python
class FlyingWizard(Wizard, FlyingMixin):
  def __init__(self, name, health, mana):
    Wizard.__init__(self, name, health, mana)
    FlyingMixin.__init__(self)

flying_wizard = FlyingWizard("Albert", 100, 20)
flying_wizard.show_status()
flying_wizard.perform_action("fireball")
flying_wizard.show_flight_status()
flying_wizard.take_off()
flying_wizard.show_flight_status()
flying_wizard.land()
flying_wizard.show_flight_status()
````
