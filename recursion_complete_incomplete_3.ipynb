## Recursion Support / Review

A recursive function is a function that calls itself.

Every recursive function should have:

1. **A base case**  
   The stopping condition. This prevents the recursion from continuing forever.

2. **A recursive case**  
   The part of the function where the function calls itself on a smaller or simpler version of the problem.

A good question to ask yourself is:

> How does this problem get smaller each time?

Another important idea is the difference between **winding** and **unwinding**.

During the **winding phase**, recursive calls are being made and stack frames are being added.

During the **unwinding phase**, base cases have been reached and return values begin traveling back up through the stack frames.

Not all recursion is inefficient. Recursion is often efficient when:

- there is approximately one recursive call per input element
- there is no recursive branching
- there is no repeated or duplicated work

In these cases, the recursive algorithm often has **O(n)** time complexity.

# Question 3 — Recursive Multiplication

Create a recursive function:

```python
multiply(a, b)
```

that multiplies two positive integers using recursive addition.

Example:

```python
multiply(4, 3)
```

should return:

```text
12
```

because:

```text
4 + 4 + 4 = 12
```

## Requirements

- Use recursion.
- Do not use the multiplication operator `*` inside your function.
- You may assume `b` is a non-negative integer.
- Include at least 3 test cases.

## Hint

Think of multiplication as repeated addition.

A possible recursive idea is:

```text
multiply(a, b) = a + multiply(a, b-1)
```

The problem gets smaller because `b` decreases.


````python
def multiply(a, b):
    # base case:
    if b == 0:
      return 0

    # recursive case:
    return a + multiply(a, b-1)



# Test cases
print(multiply(4, 3))   # expected 12
print(multiply(5, 2))   # expected 10
print(multiply(7, 0))   # expected 0
````

# Question 4 — Recursive Palindrome

Create a recursive function:

```python
is_palindrome(text)
```

that returns `True` if a string is a palindrome and `False` otherwise.

A palindrome reads the same forward and backward.

Examples:

```python
is_palindrome("racecar")
```

returns:

```python
True
```

```python
is_palindrome("python")
```

returns:

```python
False
```

## Requirements

- Use recursion.
- Do not use loops.
- Do not reverse the string using slicing, such as `text[::-1]`.
- Include at least 3 test cases.

## Hint

Compare the first and last characters.

If they do not match, the string is not a palindrome.

If they do match, recursively check the smaller string inside them.


````python
def is_palindrome(text, first=0, last=-1):
    # base case:
    if first >= len(text) // 2:   # if we've already checked the first half of the numbers
      return True

    # recursive case:
    if not text[first] == text[last]:
      return False
    return is_palindrome(text, first +1, last -1)


# Test cases
print(is_palindrome("racecar"))  # expected True
print(is_palindrome("python"))   # expected False
print(is_palindrome("level"))    # expected True
````
