```python
class Pet:
  country = "USA"
  def __init__(self, name: str, age: int, owner: str) -> None:
    self.name = name
    self.age = age
    self.owner = owner

  # Getters:
  def get_name(self):
    return self.name
  def get_age(self):
    return self.age
  def get_owner(self):
    return self.owner

  # Setters:
  def set_name(self, name: str):
    self.name = name
  def set_age(self, age: int):
    self.age = age
  def set_owner(self, owner: str):
    self.owner = owner

pet1 = Pet("Fluffy", 5, "Brenda")
pet2 = Pet("Buddy", 3, "Bob")
print(pet1.get_name())
pet2.set_name("Bracha")
print(pet2.get_name())

Pet.country = "Israel"
print(pet1.country)
print(pet2.country)

pet1.country = "France"
print(pet1.country)
print(pet2.country)
```
