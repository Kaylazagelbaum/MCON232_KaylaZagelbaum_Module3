## Exercise : to build a dictionary that looks like this
```
students_demo = {

    "Bob": {"age": 21, "major": "Math", "grades": {"Math": 85, "CS": 88}}
}
```

````python
# Your answer here
students = {}

def load_csv_data():
  with open("student_data.csv", "r") as fin:
    header = [h.strip() for h in fin.readline().strip().split(",")]
    print(header)

    for line in fin:
      parts = [p.strip() for p in line.strip().split(",")]

      if len(parts) != 5:
        continue

      name, age, major, course, grade = parts
      age = int(age)
      grade = int(grade)

      if name not in students:
        students[name] = {"age": age, "major": major, "grades": {}}

      students[name]["grades"][course] = grade


load_csv_data()
print(students)
````

## Leetcode 202 - Happy Number
Write an algorithm to determine if a number n is happy.

A happy number is a number defined by the following process:

Starting with any positive integer, replace the number by the sum of the squares of its digits.
Repeat the process until the number equals 1 (where it will stay), or it loops endlessly in a cycle which does not include 1.
Those numbers for which this process ends in 1 are happy.
Return true if n is a happy number, and false if not.



Example 1:

Input: n = 19
Output: true
Explanation:
12 + 92 = 82
82 + 22 = 68
62 + 82 = 100
12 + 02 + 02 = 1
Example 2:

Input: n = 2
Output: false


Constraints:

1 <= n <= 231 - 1

````python
# solution
# replace the positive integer with the sum of the squares of its digits\
# if the process ends in 1, it's happy, otherwise it's not

def happy_number(num):
  if not isinstance(num, int):
    return None

  seen = set()
  while num != 1 and num not in seen:
    seen.add(num)
    num = sum(int(d) ** 2 for d in str(num))
    print(num, end=",")

  return num == 1

happy_number(64)


````
