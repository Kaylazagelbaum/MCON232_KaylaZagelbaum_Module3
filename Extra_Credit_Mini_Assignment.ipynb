```python
from abc import ABC, abstractmethod
class Doctor(ABC):
  num_of_docs = 0
  def __init__(self, name, specialty):
    self.name = Doctor.format_name(name)    # uses the statics method
    self.specialty = specialty
    Doctor.num_of_docs += 1   #increments the class attribute

  @abstractmethod
  def checkup(self):
    pass

  @classmethod
  def get_num_docs(cls):
    return cls.num_of_docs

  @staticmethod
  def format_name(name):
    return "Dr. " + name.strip().lower().capitalize()


class Optometrist(Doctor):
  def __init__(self, name, specialty="optometry"):
    super().__init__(name, specialty)

  def checkup(self):
    return f"I, {self.name}, do {self.specialty} and I'm here to check your eyes."

  def sell_glasses(self):
    return "Please stop by our glasses store to pick your next pair of glasses."

class Pediatrician(Doctor):
  def __init__(self, name, specialty="pediatrics"):
    super().__init__(name, specialty)

  def checkup(self):
    return f"I, {self.name}, do {self.specialty} and I'm here to conduct your yearly well-visit."

  def give_sticker(self):
    return "Thank you for coming and remember to get stickers on your way out!"

class Dentist(Doctor):
  def __init__(self, name, specialty="dentistry"):
    super().__init__(name, specialty)

  def checkup(self):
    return f"I, {self.name}, do {self.specialty} and I'm here to check if you have any cavities."

  def reminder(self):
    return "Remember to brush your teeth twice a day and keep smiling."

opt = Optometrist("Jenny")
ped = Pediatrician("George")
den = Dentist("Alice")
docs = [opt, ped, den]
for doc in docs:
  print(doc.checkup())

print(den.get_num_docs())
```
