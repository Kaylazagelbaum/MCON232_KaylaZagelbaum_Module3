## Problem 1: Build an Animal Hierarchy

Create a parent class named `Animal`.

### The `Animal` class must:
- store a name
- store an age
- include a method named `speak()` that returns a general animal sound
- include a method named `move()` that returns a general description of movement
- include a method named `describe()` that returns a short sentence about the animal

Then create these child classes:

- `Dog`
- `Cat`
- `Rabbit`

### Child class requirements
Each child class must:
- inherit from `Animal`
- use the parent constructor appropriately
- override `speak()`
- override `move()`

### After creating your classes
Create one object of each child class and demonstrate that:
- each object stores its own data
- each object gives a different result for `speak()`
- each object gives a different result for `move()`
- `describe()` works for each object


```python
class Animal:
  def __init__(self, name: str, age: int) -> None:
    self.name = name
    self.age = age
    self.animal = "animal"
  def speak(self):
    return "Bowow"
  def move(self):
    return "crawl away"
  def describe(self):
    return f"This {self.animal} is named {self.name} and is {self.age} years old"

class Dog(Animal):
  def __init__(self, name: str, age: int) -> None:
    super().__init__(name, age)
    self.animal = "dog"
  def speak(self):
    return "Woof woof!"
  def move(self):
    return "Wag your tail!"

class Cat(Animal):
  def __init__(self, name: str, age: int) -> None:
    super().__init__(name, age)
    self.animal = "cat"
  def speak(self):
    return "Meow!"
  def move(self):
    return "Raise your ears!"

class Rabbit(Animal):
  def __init__(self, name: str, age: int) -> None:
    super().__init__(name, age)
    self.animal = "rabbit"
  def speak(self):
    return "Meep"
  def move(self):
    return "Hop on over!"

my_dog = Dog("Tyson", 5)
my_cat = Cat("Robert", 7)
my_rabbit = Rabbit("Johnny", 6)

print(f"{my_dog.name}: {my_dog.speak()}, {my_dog.move()}, {my_dog.describe()}")
print(f"{my_cat.name}: {my_cat.speak()}, {my_cat.move()}, {my_cat.describe()}")
print(f"{my_rabbit.name}: {my_rabbit.speak()}, {my_rabbit.move()}, {my_rabbit.describe()}")
```

## Problem 2: Build a Vehicle Hierarchy

Create a parent class named `Vehicle`.

### The `Vehicle` class must:
- store a brand
- store a speed
- include a method named `move()` that returns a general statement about transportation
- include a method named `sound_horn()` that returns a general horn sound
- include a method named `describe()` that returns a short sentence about the vehicle

Then create these child classes:

- `Car`
- `Truck`
- `Motorcycle`

### Child class requirements
Each child class must:
- inherit from `Vehicle`
- use the parent constructor appropriately
- override `move()`
- override `sound_horn()`

### After creating your classes
Create one object of each child class and demonstrate that:
- each object stores its own data
- each object gives a different result for `move()`
- each object gives a different result for `sound_horn()`
- `describe()` works for each object


```python
class Vehicle:
  def __init__(self, brand: str, speed: int) -> None:
    self.brand = brand
    self.speed = speed
    self.vehicle = "vehicle"
  def move(self):
    return "Drive down the street"
  def sound_horn(self):
    return "Toot toot"
  def describe(self):
    return f"I am a {self.brand} {self.vehicle} and I am driving at {self.speed} mph."

class Car(Vehicle):
  def __init__(self, brand, speed):
    super().__init__(brand, speed)
    self.vehicle = "car"
  def move(self):
    return "Turn the corner"
  def sound_horn(self):
    return "Beep Beep"

class Truck(Vehicle):
  def __init__(self, brand, speed):
    super().__init__(brand, speed)
    self.vehicle = "truck"
  def move(self):
    return "Rumble down the road"
  def sound_horn(self):
    return "Honk Honk"

class Motorcycle(Vehicle):
  def __init__(self, brand, speed):
    super().__init__(brand, speed)
    self.vehicle = "motorcycle"
  def move(self):
    return "Rev your engine"
  def sound_horn(self):
    return "Vroom Vroom"

my_car = Car("Toyota", 60)
print(f"Car: {my_car.move()}, {my_car.sound_horn()}, {my_car.describe()}")
my_truck = Truck("Ford", 70)
print(f"Truck: {my_truck.move()}, {my_truck.sound_horn()}, {my_truck.describe()}")
my_motorcycle = Motorcycle("Mercedes", 50)
print(f"Motorcycle: {my_motorcycle.move()}, {my_motorcycle.sound_horn()}, {my_motorcycle.describe()}")

```


## Problem 3: Demonstrate Polymorphism with Animals

Using your `Dog`, `Cat`, and `Rabbit` classes:

1. Create a list that stores multiple animal objects together.
2. Use a loop to go through the list.
3. For each object, call the **same methods** and show that the results differ depending on the object.

### Required methods to call in the loop
- `speak()`
- `move()`
- `describe()`

## Goal
Your output should clearly demonstrate that the program can treat all of the child objects as `Animal` objects, while each one still behaves in its own way.
