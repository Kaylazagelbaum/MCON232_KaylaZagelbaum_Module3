## Question 1: Inheritance with Employees

Create a parent class named `Employee`.

### `Employee` must:
- store an employee name
- store a department
- include a method named `work()` that returns a general work message

Then create a child class named `Manager`.

### `Manager` must:
- inherit from `Employee`
- use the parent constructor
- override `work()` so that it returns a manager-specific message

### After writing your classes
Create a `Manager` object and demonstrate that:
- the object stores its data correctly
- the overridden `work()` method is being used


````python
class Employee:
  def __init__(self, name: str, department: str):
    self.name = name
    self.department = department

  def work(self):
    return f"{self.name} works in the {self.department} department"

class Manager(Employee):
  def __init__(self, name, department):
    super().__init__(name, department)

  def work(self):
    return f"{self.name} manages the {self.department} department"

worker = Employee("Alice", "Healthcare")
print(worker.work())
boss = Manager("Martha", "Home Goods")
print(boss.work())
````


## Question 2: Choose What Belongs in the Parent

Below is a group of possible attributes and behaviors for electronic devices:

- brand
- battery_percent
- charge()
- take_photo()
- make_call()

Create a parent class named `Device` using only the information that should reasonably be shared by many devices.

Then create a child class named `Phone`.

### Requirements
- `Device` should store shared data and include at least one shared behavior
- `Phone` should inherit from `Device`
- `Phone` should add or override behavior that makes sense for a phone

### After writing your classes
Create a `Phone` object and demonstrate that inherited behavior works correctly.


````python
class Device:
  def __init__(self, brand: str, battery_percent: int):
    self.brand = brand
    self.battery_percent = battery_percent

  def charge(self, percent):
    self.battery_percent += percent
    return "Device has been charged with a charger. New battery is:", self.battery_percent

class Phone(Device):
  def __init__(self, brand, battery_percent):
    super().__init__(brand, battery_percent)

  def charge(self, percent):
    self.battery_percent += percent
    return f"This phone has been charged with a universal-c charger. The new battery is {self.battery_percent}%"

iPhone = Phone("Apple", 70)
print(iPhone.charge(10))
````


## Question 3: Polymorphism with School Roles

Create a parent class named `SchoolMember` with a method named `introduce()`.

Then create these child classes:
- `Teacher`
- `Student`

Each child class should override `introduce()` in its own way.

### Then:
1. Create one `Teacher` object and one `Student` object
2. Store them together in a list
3. Use one loop to call `introduce()` on each object

## Goal
Your output should clearly demonstrate polymorphism.


````python
class SchoolMember:
  def __init__(self, name):
    self.name = name

  def introduce(self):
    return f"Hi, my name is {self.name}."

class Teacher(SchoolMember):
  def __init__(self, name):
    super().__init__(name)

  def introduce(self):
    return f"Hi, my name is {self.name} and I am so excited to teach you this year."

class Student(SchoolMember):
  def __init__(self, name):
    super().__init__(name)

  def introduce(self):
    return f"Hi, my name is {self.name} and I am so excited to be a student in this class!"

Alice = Student("Alice")
Betsy = Student("Betsy")
Mrs_Adams = Teacher("Mrs. Adams")

school_members = [Alice, Betsy, Mrs_Adams]

for member in school_members:
  print(member.introduce())
````

## Bonus (Optional)

Create your own small inheritance example with:
- 1 parent class
- 2 child classes
- 1 overridden method
- 1 loop that demonstrates polymorphism


````python
class Shape:
  def __init__(self, length, width):
    self.length = length
    self.width = width

  def area(self):
    return "Find the area of a shape."

class Square(Shape):
  def __init__(self, length, width):
    super().__init__(length, width)

  def area(self):
    return self.length * self.width

class R_Triangle(Shape):
  def __init__(self, length, width):
    super().__init__(length, width)

  def area(self):
    return .5 * self.length * self.width

square1 = Square(5, 10)
square2 = Square(6, 20)
right_triangle = R_Triangle(4, 14)

shapes = [square1, square2, right_triangle]
for shape in shapes:
  print(shape.area())

````
